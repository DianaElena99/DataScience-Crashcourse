# Matplotlib Crashcourse

#Plot a line

```python
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-3,3,0.001)
plt.plot(x, st.norm.pdf(x))
plt.show() 
```

# Multiple plots on one graph

```python
plt.plot(x, st.norm.pdf(x))
plt.plot(x, st.norm.pdf(x, 2.0, 0.5))
plt.show() 
```

# Adjust the axes

```python
axes = plt.axes()
axes.set_xlim([-3,3])
axes.set_ylim([0,1])
axes.set_xticks([-3,-2,-1,0,1,2,3])
axes.set_yticks([0,0.2,0.4,0.6,0.8])
axes.grid()
plt.plot(x, st.norm.pdf(x))
plt.plot(x, st.norm.pdf(x, 2.0, 0.5))
plt.show()
```

# Labels and Legends

```python
axes = plt.axes()
axes.set_xlim([-3,3])
axes.set_ylim([0,1])
axes.set_xticks([-3,-2,-1,0,1,2,3])
axes.set_yticks([0,0.2,0.4,0.6,0.8])
axes.grid()

plt.xlabel('Index')
plt.ylabel('Probability')

plt.plot(x, st.norm.pdf(x))
plt.plot(x, st.norm.pdf(x, 2.0, 0.5))

plt.legend(['Plot 1', 'Plot 2'], loc=2)

plt.show()
```

# Pie Charts

```python
data = [12,55,24,65,48]
colors = ['b','g','r','c','m']
explode = [0.2, 0, 0, 0.1, 0]
labels = ['Anthropology', 'History', 'Psychology', 'Philosophy', 'Economics']
title = 'Books by Category'
plt.pie(data, colors = colors, labels=labels, explode = explode)
plt.title(title)
plt.show()
```

# Bar Charts

```python
avg_heights = [151.6, 155.23, 160.8, 165.66]
ages = [12, 14, 16, 19]
axes = plt.axes()
axes.set_ylim([145,170])
axes.set_yticks([145,150,155,160,165,170])
plt.bar(ages, avg_heights)
plt.show()
```
